```python
import pandas as pd

url = "https://catalog.ourworldindata.org/garden/covid/latest/compact/compact.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.dropna(axis=1, thresh=len(df)*0.9)
df = df.dropna(subset = ['country','date'])

df['new_cases'] = df['new_cases'].fillna(0)
df['total_cases'] = df['total_cases'].fillna(1)


```

```python
india_data = df[df['country'] == 'India']

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(india_data['date'], india_data['new_cases'], label='New Cases per Day')
plt.title('Daily COVID-19 Cases in India')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
